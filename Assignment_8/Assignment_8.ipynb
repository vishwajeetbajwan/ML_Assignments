## Multiple Linear Regression

### Loading the data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'data\kc_house_data.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
data.columns
```

```python
cols = ['bedrooms','bathrooms','sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated','sqft_living15', 'sqft_lot15']
# these columns will be used for the prediction
```

```python
df = data[cols] #creating new dataframe without 'price' column for prediction
```

```python
df.head()
```

### Feature Transform

```python
# standardization
from sklearn.preprocessing import StandardScaler
StandardScaler().fit_transform(df)
```

```python
df.columns = cols
df.head()
```

### Preparing the Data - Test Train

```python
x = df[cols]
y = data['price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x , y, test_size = 0.3, random_state = 0)
```

### Training

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

### Making the Prediction

```python
y_pred = regressor.predict(x_test)
```

```python
df = pd.DataFrame({'Actual': y_test.astype(int), 'Predicted': y_pred.astype(int)})
df.head(10)
```

### Evaluating the Algorithm

```python
from sklearn import metrics

print('Mean Absolute Error: ', metrics.mean_absolute_error(y_test, y_pred))

print('Mean Squared Error: ', metrics.mean_squared_error(y_test, y_pred))

print('Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
