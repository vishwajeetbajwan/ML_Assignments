```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data\weather.csv")
df = pd.DataFrame(data)
df.head()
```

```python
x = df["MinTemp"]
```

```python
sns.distplot(x)
```

```python
sns.distplot(x,kde=False,rug=True)
```

```python
sns.distplot(x,hist=False,rug=True)
```

```python
sns.kdeplot(x,shade=True)
```

```python
sns.jointplot(x="Evaporation",y="Sunshine",data=df)
```

```python
with sns.axes_style("white"):
    sns.jointplot(x=df['Pressure3pm'], y=df["Humidity3pm"], kind="hex",color="k")
```

```python
sns.jointplot(x=df["Pressure9am"],y=df["Temp9am"],kind="kde")
```

```python
df1 = pd.DataFrame(df, columns=['Humidity9am','Humidity3pm','Pressure9am','Pressure3pm'])
sns.pairplot(df1)
```

```python
df1 = pd.DataFrame(df, columns=['Humidity9am','Humidity3pm','Pressure9am','Pressure3pm','RainToday'])
sns.pairplot(df1,hue='RainToday')
```

```python
sns.stripplot(x=df['Cloud9am'],y=df['Humidity9am'])
```

```python
sns.stripplot(x=df['Cloud9am'],y=df['Humidity9am'],jitter=True)
```

```python
sns.swarmplot(x=df['Cloud9am'],y=df['Humidity9am'])
```

```python
sns.boxplot(x=df['Cloud9am'],y=df['Humidity9am'])
```

```python
sns.boxplot(x=df['Cloud9am'], y=df['Humidity9am'], data=df, hue='RainToday', dodge=True)
```

```python
sns.barplot(x=df['Temp9am'],y=df['Humidity9am'])
```

```python
sns.barplot(x=df['Cloud9am'],y=df['Humidity9am'])
```

```python
sns.barplot(x=df['Cloud9am'],y=df['Humidity9am'], data=df, hue='RainToday')
```

```python
sns.countplot(x='RainToday', data=df, palette='Greens_d')
```

```python
sns.countplot(x='RainTomorrow', data=df, palette='Greens_d')
```
